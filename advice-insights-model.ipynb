```python
model = genai.GenerativeModel("gemini-1.5-pro-latest")
```

```python
def analyze_interface(text, language='en'):
    formatted_comments = text.strip()

    if language == 'en':
        prompt = f"""
        You have some reviews with ratings about a service. Your task is to find the main issues that customers are mentioning, calculate the overall success rate, and provide simple and quick advice to the service provider.
        
        **Reviews:**
        {formatted_comments}
        **Main Issues:**
        - 
        - 
        - 
        **Success Rate:**
        - 
        - 
        **Advice:**
        - 
        - 
        - 
        """
    else:
        prompt = f"""
        عندك شوية تعليقات بتقييمات عن خدمة، طلع المشاكل المتكررة، نسبة النجاح، ونصيحة لصاحب الخدمة بأسلوب بسيط وسريع.
        **التعليقات:**
        {formatted_comments}
        **أهم المشاكل:**
        - 
        - 
        - 
        **نسبة النجاح:**
        ** النصيحه المقدمه لك :-:**
        - 
        - 
        - 
        """
    
    try:
        response = model.generate_content(prompt)
        return response.text.strip()
    except Exception as e:
        return f"حصل خطأ: {str(e)}"

```

```python
with gr.Blocks() as demo:
    gr.Markdown("# تحليل التقييمات")
    gr.Markdown("اكتب تعليقات العملاء مع تقييماتهم زي ⭐ 4/5: الخدمة جيدة، واضغط تحليل")
    
    with gr.Row():
        input_text = gr.Textbox(label="التعليقات (كل تعليق في سطر)")
        language_radio = gr.Radio(["en", "ar"], label="اختر اللغة", value="en")
    
    output_text = gr.Textbox(label="التحليل")
    
    submit_button = gr.Button("تحليل")
    submit_button.click(analyze_interface, inputs=[input_text, language_radio], outputs=output_text)


demo.launch()
```
